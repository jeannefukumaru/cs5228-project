```python
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
@slicing_function()
def cluster_label_0(x):
    sentence = model.encode(x.text)
    label = tweets_kmeans.predict(sentence)
    return bool(label == 0)
```
